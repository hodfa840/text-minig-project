# Pie chart and Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the number of occurrences of each label
class_counts = data['class'].value_counts()

# Create a pie chart to retrieve colors
plt.figure(figsize=(8,8))
pie = plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
colors = [x.get_facecolor() for x in pie[0]]
plt.legend(class_counts.index, bbox_to_anchor=(1.55, 0))

# Create a bar plot using same colors
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
ax = sns.barplot(x=class_counts.index, y=class_counts.values, palette=colors)

# Set axis labels and title
ax.set(xlabel='Classes', ylabel='Counts', title='Distribution of Classes')

# Rotate x-axis labels for better visibility
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

# Show the bar plot
plt.show()
```

From the pie chart, it can be observed that the data is distributed almost equally.
